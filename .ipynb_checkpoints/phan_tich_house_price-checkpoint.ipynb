```python
import pandas as pd
import numpy as np

data = pd.read_csv('house_price_Dống-Da_Hà-Nội_subdata.csv')
data
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
```

```python
# Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên
# ==> Dữ liệu không có thuộc tính phường
```

```python
# Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công
# của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.
data3 = data[(data['land_certificate']=='So do')&(data['bedroom']>=3)].loc[:,['price','house_direction','balcony_direction']]
data3
```

```python
# Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.
data4 = pd.pivot_table(data,values='price',index=['property_type'],aggfunc=['mean','min','max'])
data4
```

```python
# Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.
# ==> Dữ liệu không có thuộc tính phường, tạm thay bằng property_type
data5 = pd.pivot_table(data,values=['bedroom','toilet','floor'],index=['property_type'],aggfunc='mean')
data5
```
