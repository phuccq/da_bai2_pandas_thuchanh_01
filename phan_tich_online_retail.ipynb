```python
import pandas as pd

data = pd.read_csv('OnlineRetail.csv')
data
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
```

```python
# Tạo cột mới có tên quý –  ‘Previous’:
# nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3;
# nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6;
# nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9; 
# nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12
import numpy as np

invoice_month = pd.to_datetime(data['InvoiceDate']).dt.month
# invoice_month.values
# data['Previous'] = np.where(data['invoice_month'] in ('1','2','3'),1,2)
conditions = [invoice_month.isin([1,3]),\
              invoice_month.isin([4,6]),\
             invoice_month.isin([7,9])]
choices = [1,2,3]
data['Previous'] = np.select(conditions,choices,default=4)
# data['Previous'].value_counts()
```

```python
# Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
data['Amount'] = data['Quantity']*data['UnitPrice']
```

```python
# Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’
# và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.
discount_condition = [(data['Country']=='United Kingdom') & (data['Previous']==4),data['Country']=='France']
discount_choice = ['10%','5%']
data['Discount'] = np.select(discount_condition,discount_choice,default='0%')
# data['Discount'].value_counts()
```

```python
# Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.
data['Total'] = (1-float(data['Discount'].values[0].strip('%'))/100) *data['Amount']
data
```
