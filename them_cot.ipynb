```python
import pandas as pd

data = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv')
data = data[['join_month', 'join_day','join_year','shop_location','rating_bad','rating_good','rating_normal']]
data
```

```python
# Tạo cột 'rating' tính điểm cho mỗi cửa hàng dựa vào thông tin các cột rating_bad, rating_good, rating_normal 
# theo công thức sau: rating = rating_good *2 + rating_normal - rating_bad*3
data['rating']=data['rating_good']*2+data['rating_normal']-data['rating_bad']*3
data
```

```python
# ghép 3 cột join_month, join_day, join_year thành cột mới có tên 'date' nhận giá trị có dạng:
# "join_month join_day,join_year"
# dùng hàm astype(str) để đổi một giá trị có kiểu bất kỳ về kiểu chuỗi.
data['date']=data['join_month']+' '+data['join_day'].astype(str)+' '+data['join_year'].astype(str)
data
```

```python
# Thêm cột new có giá trị True nếu join_year = 2021 và False trong trường hợp còn lại.
data['new']=data['join_year']==2021
data
```

```python
# Nếu có 2 lựa chọn ta sử dụng np.where
# Thêm cột rate có giá trị good nếu rating_good >= 50000,  bad trong trường hợp còn lại
import numpy as np
data['rate']=np.where(data['rating_good']>=50000,'good','bad')
data
```

```python
# Nếu có nhiều hơn 2 lựa chọn ta sử dụng np.select
# Thêm cột flag tặng cờ cho các cửa hàng, flag nhận các giá trị như sau:
# blue khi rating_good >= 30000 và rating_bad <= 100
# yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
# red khi rating_good < 10000
# black đối với các trường hợp còn lại

conditions = [(data['rating_good']>=30000) & (data['rating_bad']<=100),
              (data['rating_good']<30000) & (data['rating_good']>=10000) & (data['rating_bad']>100) & (data['rating_bad']<=1000),
              (data['rating_good']<10000)]
choices = ['blue','yellow','red']
data['flag']=np.select(conditions,choices,default='black')
data
```
